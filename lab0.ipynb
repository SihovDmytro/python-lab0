```python
# У введеному списку залишити тільки непрості числа і відсортувати їх за зменшенням
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def task4():
    text = input()
    tokens = text.split(" ")
    nonPrimeNumbers = []
    invalid = []
    for token in tokens:
        token = str(token)
        if(token.isnumeric()):
            if(not isPrime(int(token))):
                nonPrimeNumbers.append(int(token))
            else:
                print("Просте число пропущено:", token)
        else:
            print("Невалідне значення пропущено: ", token)
            invalid.append(token)
    
    nonPrimeNumbers.sort(reverse=True)
    return nonPrimeNumbers

# 10 2 31 42 5 62 str 7 18 9 55 11 20 20 30 15 16 ddd123 17 18 19
# 100 1abrakadabra2 101 200 7 a 5 b 9 c 6 12
print("Результат:", task4())
```
